# Who eats the Food we Grow?

- Food: refers to the total amount of the food item available as human food during the reference period.

- Feed: refers to the quantity of the food item available for feeding to the livestock and poultry during the reference period.

El dataset es demasiado grande, con 21477 filas, y 63 columnas. Asi que voy simplificar mi analisis, al extraer un subconjunto del dataset: solo para Norteamerica, y solo la decada de los 2000's

```python
import pandas as pd
import numpy as np

df_food = pd.read_csv('../Data_Base/FAO.csv', sep=',', encoding='ISO-8859-1')
df_food.shape
```

```python
df_food.head(2)
```

Usando Filtrado por condiciones, para obtener un Data Frame de solo los paises de NorteAmerica

```python
df_food = df_food[(df_food.Area == 'Mexico') | (df_food.Area == 'Canada') | (df_food.Area == 'United States of America')]
```

```python
df_food.groupby('Area').count()
```

```python
df_food = df_food.drop(df_food.iloc[:,10:49], axis=1)
df_food = df_food.drop(df_food.iloc[:, -4:], axis=1)
df_food.shape
```

Finalmente, se obtiene un dataset solo con la informacion relevante, del aÃ±o 2000 al 2009, y paises de America del Norte

```python
df_food.head(2)
```

Agregando una Columna que contiene la suma de las columnas desde el Y2000 al Y2009

```python
df_food['Suma'] = df_food.iloc[:, 10:].sum(axis=1)
df_food.head(2)
```

Tener Cuidado con esta columna *Suma*, porque como demostraremos al hacer un simple agrupamiento, no coincide si se suman todos los aÃ±os desde 2000 al 2009

```python
df_food.groupby('Element').sum()
```

```python
df_aux = df_food.groupby('Element').sum()
```

Los valores obtenidos a continuacion son diferentes a los de la Columna *Suma*. Obtengo el mismo resultado con Excel, lo que indica esta bien.

```python
df_aux.iloc[:,-11:-1].sum(axis=1)
```

Mejor borremos la columana *Suma* y cada vez que necesite calcular las uma por alguna razon, la evaluo ahi mismo

```python
df_food.drop('Suma', axis=1, inplace=True)
```

```python
df_food.head(2)
```

Realizo un agrupamiento por Pais: por ejemplo, Canada tiene 140 registros, y para el aÃ±o 2000, habia disponible 801 T de comida tanto para consumo humano como para alimentar animales. 

```python
df_food.groupby('Area').agg(['count', 'mean']).iloc[:,10:]
```

Mediante una funcion de agrupamiento doble, logro ilustrar que tanta candidad de comida estaba disponible por pais, y por elemento: para consumo animal o humano. Se reduce la carga cognitiva, y creo la table es muy clara

```python
df_food.groupby(['Area', 'Element']).sum().iloc[:,5:]
```
