## Filtrado con LOC and ILOC

Cuando queremos navegar por un dataFrame estas funciones permiten filtrar datos de manera más específica

```python
import pandas as pd

best_sellers = pd.read_csv('../Data_Base/bestsellers_with_categories.csv', sep=',', header=0)
best_sellers.head()
```

Hagamos un slicing de los primeros 10 registros

```python
best_sellers = best_sellers[:10]
best_sellers
```

Filtremos por columnas, donde solo aparezcan Name and Author. Primero por Nombre

```python
best_sellers['Name']
```

Ahora si por los dos

```python
best_sellers[['Name', 'Author']]
```

### LOC 

Para que muestre todo

```python
best_sellers.loc[:]
```

Para filtrar del 0 al 4 incluyendolo

```python
best_sellers.loc[0:4]
```

Vemos que tambien podemos usar *slicing* para que muestre lo mismo pero funciona un poco diferente a *loc*

```python
best_sellers[0:5]
```

#### QUe mas se puede hacer con *loc*

```python
best_sellers.loc[0:4, ['User Rating']] * (-1)
```

Ahora quiero comparar en todo el *data frame* cuando el autor es *Stephen King*

```python
best_sellers.loc[:, ['Author']] == 'Stephen King'
```

### iloc

Es muy parecido a loc. Lo que ocurre es que ahora la busqueda va a ser por indice, de hay el prefijo 'i'.

En el siguiente ejemplo le estoy diciendo que quiero todas las filas, pero las columnas del 0 a la 2  


```python
best_sellers.iloc[:, 0:2]
```

En el siguiente ejemplo voy a comparar donde el autor sea *Amor Towels*

```python
best_sellers.iloc[:,1] ==  'Amor Towles'
```

O muy util para filtrar mucho mas los datos. Y una manera mucho mas sencilla para navegar entre todos estos datos

```python
best_sellers.iloc[2:5, 1:4]
```

## Manjeo de Datos Nulos

Los datos nulos son un dolor de cabeza en el mundo de Data Science, y aqui vamos a ver como tratrar con ellos.

Como punto de partida, se crea un dicccionario, y a partir de el se crea un objeto de tipo DataFrame

```python
import pandas as pd
import numpy as np

my_dict = {'Col1':[1,2,3,np.nan],
'Col2': [4, np.nan, 6,7],
'Col3': ['a','b','c', None]}

df = pd.DataFrame(my_dict) 
df.head()
```

El dataset que acabamos de crear es muy comun verlo, con valores incompletos y nulos. El primer metodo seria *is_null*, que retorna True si es un valor None or NaN

```python
df.isnull()
```

La mejor forma de hacerlo es multiplicaarlo por 1, y ya veras porque

```python
df.isnull()*1
```

Una vez los hallamos identifado, y para decirlo de alguna forma, normalizar los datos, colocarles una etiqueta de 'Missing' con el metodo *fillna*

```python
df.fillna('Missing')
```

Pero eso no es todo, puedo usar *fillna* y llenarlos con valores numericos, por ejemplo la media, la moda, la STV, etc

Para este caso me indicara un warning porque la columna 3 obviamente no son valores numericos

```python
df.fillna(df.mean())
```

Otra forma de tratar con valores NaN es interpolarlos, es decir, se le indicara que es una serie de datos, y tratara de buscar el numero que hace falta para completar la serie: *interpolate*

```python
df.interpolate() 
```

Y una ultima forma de tratar con los NaN, y muy comun entre los cientificos de datos, es eliminarlos: *dropna*. Esto como siempre es usando el mejor criterio. Si por ejemplo son tan pocos los valores NaN con relacion al tamaño del DataFrame, pues mejor borrarlos 

```python
df.dropna()
```

## Filtrado por Condiciones

Al igual que en Numpy, tambien podemos filtrar en Pandas

```python
import pandas as pd

df = pd.read_csv('../Data_Base/bestsellers_with_categories.csv')
print(df.shape)
df.head()
```

Por ejemplo donde al año sea mayor o igual a 2015

```python
df['Year'] >= 2017
```

Como ves es muy parecido a como lo veniamos haciendo con Numpy, porque Pandas esta basado en Numpy.

Usemos la condicion anterior para filtrar los datos

```python
mayor_igual_2017 = df['Year'] >= 2017

df[mayor_igual_2017]
```

Y como ves ahora tenemos 150 registros a comparacion de los 550 originales. Tambien se puede implementar directamente...

```python
df[df['Year'] >= 2017]
```

#### Ejemplo

Filtrar los datos de tal forma que solo aparezcan los que no sean de ciencia ficcion

```python
df[df['Genre'] != 'Fiction']
```

Y tambien, sumando complejidad integrando mas condiciones.

#### Ejemplo 

Encontrar todos los registros donde el genero se Fiction, y el año de publicacion mayor a 2016. Encontrando 65 filas que cumplen con la condicion

```python
fiction = df['Genre'] == 'Fiction'

mayor_2016 = df['Year'] > 2016


df[fiction & mayor_2016]
```

Si queremos negar la condicion, y para reutilizar codigo, usamos el simbolo **~** para negarla. POr ejemplo encontremos los libros de Ficcion que no sean mayores a 2016.  

```python
df[fiction & ~mayor_2016]
```

## Funciones Principales de Pandas

Hay ciertas funciones que son muy importantes y que siempre estaremos usando a la hora de hacer análisis de datos, para mayor facilidad y comprensión del DataFrame. Veremos las mas comunes:

- metodo head().
- metodo info(). Mostrar los diferentes datos que contiene el DataFrame
- metodo describe(). Obtener diferentes datos estadísticos de las columnas numéricas.
- metodo tail(). muestra los ultimos 5 registros por defecto
- memory_usage(). Obtener el uso de la memoria de cada columna
- metodo values_counts(): Obtener cuantos datos tenemos de algo en específico
- metodo drop_duplicates(): borrar registros repetidos
- metodo sort_values(): Ordenar los registros según valores de la columna (orden ascendente)

```python
import pandas as pd

df = pd.read_csv('../Data_Base/bestsellers_with_categories.csv', sep=',', header=0)
print(df.shape)
df.head()
```
