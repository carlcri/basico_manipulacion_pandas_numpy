## JOIN

Merge y COncat que los vimos anteriormente nos dan mucha velocidad. Ahora si vamos hablar de JOIN, que realiza un *index match*. HAciendo la operacion sobre los indices

```python
import pandas as pd

my_dict ={'A':['A0', 'A1','A2', 'A3'],
'B': ['B0', 'B1', 'B2', 'B3'],
'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']}

izq = pd.DataFrame(my_dict, index=['k0', 'k1', 'k2', 'k3'])
izq.head()
```

```python

my_dict = {'E':['A4', 'A5', 'A6', 'A7'],
'F': ['B4', 'B5', 'B6', 'B7'],
'G': ['C4', 'C5', 'C6', 'C7'],
'H': ['D4', 'D5', 'D6', 'D7']}

der= pd.DataFrame(my_dict, index=['k0', 'k2', 'k3', 'k5'])
der.head()
```

### LEFT JOIN

```python
izq.join(der)
```

### Inner JOin

```python
izq.join(der, how='inner')
```

### Right JOin

```python
izq.join(der, how='right')
```

### Outer Join - UNION

```python
izq.join(der, how='outer')
```

## Pivot y Melt

Son dos funciones que sirven para cambiar la estructura de nuestro DataFrame de acuerdo a nuestras necesidades.

### Pivot Table

Esta función puede traer recuerdos a las personas interesadas en el mundo del SQL, ya que Oracle, PostgreSQL y otros motores de bases de datos la tienen implementada desde hace muchos años. Pivot, básicamente, transforma los valores de determinadas columnas o filas en los índices de un nuevo DataFrame, y la intersección de estos es el valor resultante.

Todo queda mas claro con un ejemplo

```python
import pandas as pd

df_books = pd.read_csv('../Data_Base/bestsellers_with_categories.csv',sep=',',header=0)
df_books.head(2)
```

```python
df_books.pivot_table(index='Author',columns='Genre',values='User Rating')
```

Como resultado, los valores de Author pasan a formar el índice por fila y los valores de Genre pasan a formar parte de los índices por columna, y el User Rating se mantiene como valor.

Observa los valores NaN que aparecen, esto es porque un autor no puede ser Fiction y NoN-Fiction al mismo tiempo.

Por supuesto, para este caso, un Author suele tener un solo género literario, así que no es una transformación muy útil, pero veamos si podemos lograr algo mejor.

```python
df_books.pivot_table(index='Genre',columns='Year', values='User Rating',aggfunc='sum')
```

En este caso tenemos por cada género, la suma a lo largo de los años. Esto es mucho más interesante, ¿verdad? La mejor noticia es que no solo podemos obtener la suma, también podemos obtener la media, la desviación estándar, el conteo, la varianza, etc. Únicamente con cambiar el parámetro aggfunc que traduce función de agrupamiento.

- sum
- mean
- max
- min
- median
- count
- std

### MELT

El método melt toma las columnas del DataFrame y las pasa a filas, con dos nuevas columnas para especificar la antigua columna y el valor que traía.

Por ejemplo, simplemente al imprimir las cinco primeras filas del DataFrame con las columnas de Name y Genre se tiene este resultado.

```python
df_books[['Name','Genre']].head()
```

Aplica melt de la siguiente manera:

```python
df_books[['Name','Genre']].head().melt()
```

Ahora cada resultado de las dos columnas pasa a una fila de este modo a tipo llave:valor. Por ejemplo, "10-Day Green Smoothie Cleanse" que es un Name, y genero Non Fiction, despues de aplicar el *melt* la informacion esta contenido en el indice 0 e indice 5. 

```python
df_books.melt(id_vars='Year',value_vars='Genre')
```

Aplicaremos nuevamente *melt* para observar si de verdad tiene alguna utilidad practica

```python
df_books.melt(id_vars='Year',value_vars='Genre')
```

Simplemente, podemos seleccionar las columnas que no quiero hacer melt usando el parámetro id_vars. Para este caso Year y también la única columna que quiero aplicar el melt, para este caso Genre con la propiedad value_vars.

## Apply

Es un comando muy poderoso que nos deja aplicar funciones a nuestro DataFrame

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html

```python
import pandas as pd

df_books = pd.read_csv('../Data_Base/bestsellers_with_categories.csv',sep=',',header=0)
df_books.head(2)
```

Definamos una funcion

```python
def two_times(value):
    return value * 2
```

A continuacion, usamos dentro de *apply* la funcion two_times, sin hacerle un *call* esto es sin ()

Tomara los valores de la columna User Rating y los multiplicara por dos

```python
df_books['User Rating'].apply(two_times)
```

Esto es muy eficiente, mucho mas que aplicar por ejemplo un *for*

Vamos añardir una nueva columan al Data Frame con el resultado de apply

```python
df_books['Rating 2'] = df_books['User Rating'].apply(two_times)
```

```python
df_books.head()
```

### Funciones Lambda y Apply

Sobrescribiendo la columan Rating_2 con los valores de la funcion lambda, que multiplica el User Rating por 3

```python
df_books['Rating 2'] = df_books['User Rating'].apply(lambda x: x*3)
```

```python
df_books[['Name', 'Author', 'User Rating', 'Rating 2']].head()
```

Hasta ahora hacemos operaciones sobre los datos de una columan. Pero que pasaria si quisieramos hacerlo sobre varias columnas y usando lambdas

Entonces, se va usar el apply no sobre una columan en especifico, sino al Data Frame en general. En este caso el x dentro del lambda tiene el valor de todas las filas del dataframe, alrededor de todas las columnas 

```python
df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] != x['Non Fiction'] else x['User Rating'])
```

KeyError no entiendo que quiere decir 'Genre', ¿Y entonces a que estamos haciendo referencia? a las columnas, y las columnas son axis 1

```python
df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] != 'Non Fiction' else x['User Rating'], axis=1)
```

```python
df_books['Rating 2'] = df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] != 'Non Fiction' else x['User Rating'], axis=1)
```

```python
df_books[['Name', 'Genre', 'User Rating', 'Rating 2']]
```

Y como vemos, siempre y cuando el genero sea diferente de Non Ficiton, el Rating 2 lo multiplicara por 2
